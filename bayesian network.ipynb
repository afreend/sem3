# PROTOPUNCIA LIBRARY

```python
from protopunica import*

 

```

protopunica is a python library that implements probilistic models

```python
smoking = Node(DiscreteDistribution({"High smoking":0.7,"Low smoking":0.3}),name="smoking")

```

smoking is root node and is not having any parent nodes smoking is a random  variables that takes two values high smoking
and low smoking

```python
asbes_consum = Node(DiscreteDistribution({"High cons":0.3,"Low cons":0.7}),name="asbes_consum")
```

abestors consumption is also a root which is not having any parents and has two values high consumption and 
low consumption

```python
cancer = Node(ConditionalProbabilityTable([
    ["High smoking","High cons","POS",0.4],
    ["High smoking","High cons","Neg",0.6],
    ["High smoking","Low cons","POS",0.3],
    ["High smoking","Low cons","Neg",0.7],
    ["Low smoking","Low cons","POS",0.1],
    ["Low smoking","Low cons","Neg",0.9],
    ["Low smoking","High cons","POS",0.02],
    ["Low smoking","High cons","Neg",0.98],
],[smoking.distribution,asbes_consum.distribution]),
              name="cancer")
```

cancer is a random variable which conditionally depended on smokinjg and asbestos cancer takes two values positive and negative

```python
scan=Node(ConditionalProbabilityTable([
    ["POS","scan_Pos",0.8],
    ["POS","scan_Neg",0.2],
    ["Neg","scan_Pos",0.1],
    ["Neg","scan_Neg",0.9],
],[cancer.distribution]),name="scan")
```

scan(X-ray)is conditionaly dependent on canmcer.if there is a chance that the X-ray might revel it.it
takes two values positive and negative

```python
Blood_vomiting=Node(ConditionalProbabilityTable([
    ["POS","B.V_Pos",0.7],
    ["POS","B.V_Neg",0.3],
    ["Neg","B.V_Pos",0.2],
    ["Neg","B.V_Neg",0.8],
],[cancer.distribution]),name="Blood_vomiting")
```

patient suffers from blood vomiting takes two values B.vops(yes he has blood vomiting)andB.V neg(No he is not having any
blood vomiting)

# Model the bayesian network

```python
model=BayesianNetwork()
```

# Add all the nodes for the bayesian network

```python
model.add_states(smoking,asbes_consum,cancer,scan,Blood_vomiting)
```

```python
model.add_edge(smoking,cancer)
model.add_edge(asbes_consum,cancer)
model.add_edge(cancer,scan)
model.add_edge(cancer,Blood_vomiting)

```

```python
model.bake()

```

```python
model
```

```python
probability=model.probability([["Low smoking","Low cons","POS","scan_Pos","B.V_Pos"]])
```

```python
probability
```

```python
print(model.predict([["Low smoking","Low cons","POS","scan_Pos","B.V_Pos"]]))
    
```
